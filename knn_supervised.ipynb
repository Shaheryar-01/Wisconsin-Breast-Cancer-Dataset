```python
X = df.iloc[:, 1:]#all rows, all except first column
y = df.iloc[:, 0]#all rows, first column only
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
#random state ensures same chunks of data are selected always for X_train, X_Test, y_train, y_test
#ideal number for random state is 42
```

```python
len(X_train)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train) #scaler learns pattern
X_test = scaler.transform(X_test)
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = knn.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
scores = []
for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))

```

```python
np.argmax(scores) 
#shows index of max number, hence best value of k

```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 10), scores)
```
